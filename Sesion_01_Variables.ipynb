```python
resultado = input("Quieres realizar una suma o una resta?")
if(resultado == "suma"):
  print(num1 + num2)
elif(resultado == "resta"):
  print(num1 - num2)
elif(resultado == "multiplicar"):
  print(num1 * num2)
elif(resultado == "dividir"):
  print(num1 / num2)
else:
  print("Esta opcion no esta registrada")  
```

```python
print("Esta es una calculadora")
operacion = input("Que tipo de operacion quieres realizar? \n1.-Suma \n2.-Resta \n3.-Division \n4.-Multiplicacion \n")
num1 = float(input("Introduce el primer numero "))
num2 = float(input("Introduce el segundo numero "))
if operacion == "suma" or operacion == "Suma" or operacion == "SUMA" or operacion == "1":
  print("{} + {} = {}".format(num1,num2,num1+num2))
elif(operacion == "resta" or operacion == "Resta" or operacion == "RESTA" or operacion == "2"):
  print("{} - {} = {}".format(num1,num2,num1-num2))
elif(operacion == "division" or operacion == "Division" or operacion == "Division" or operacion == "3"):
  print("{} / {} = {}".format(num1,num2,num1/num2))
elif(operacion == "multiplicacion" or operacion == "Multiplicacion" or operacion == "MULTIPLICACION" or operacion == "4"):
  print("{} * {} = {} ".format(num1,num2,num1*num2))
else:
  print("Esta opcion no esta registrada")  
```
